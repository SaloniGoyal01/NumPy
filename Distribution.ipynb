## Normal (Gaussian) Distribution

```python
# Normal Distribution
# The Normal Distribution is one of the most important distributions.
# It is also called the Gaussian Distribution after the German mathematician Carl Friedrich Gauss.
# It fits the probability distribution of many events, eg. IQ Scores, Heartbeat etc.
# Use the random.normal() method to get a Normal Data Distribution.

# It has three parameters:
# loc - (Mean) where the peak of the bell exists.
# scale - (Standard Deviation) how flat the graph distribution should be.
# size - The shape of the returned array.
```

```python
from numpy import random

#  Generate a random normal distribution of size 2x3:
arr = random.normal(size = (2,3))
print(arr)
```

```python
#  Generate a random normal distribution of size 2x3 with mean at 1 and standard deviation of 2:

arr = random.normal(loc = 1, scale = 2, size = (2, 3))
print(arr)
```

##  Visualization of Normal Distribution

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.normal(size = 1000), kind="kde")
plt.show()
```

##  Binomial Distribution


```python
# Binomial Distribution is a Discrete Distribution.
# It describes the outcome of binary scenarios, e.g. toss of a coin, it will either be head or tails.

# It has three parameters:
# n - number of trials.
# p - probability of occurence of each trial (e.g. for toss of a coin 0.5 each).
# size - The shape of the returned array.
```

```python
from numpy import random

#  Given 10 trials for coin toss generate 10 data points:]
arr = random.binomial(n=10, p=0.5, size=10)
print(arr)
```

##  Visualization of Binomial Distribution

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.binomial(n=10, p=0.5, size=1000))    # size=1000 means you are generating 1000 random experiments of the binomial distribution.
plt.show()
```

# Difference Between Normal and Binomial Distribution

##### The main difference is that normal distribution is continous whereas binomial is discrete, but if there are enough data points it will be quite similar to normal distribution with certain loc and scale.
| Feature                 | **Binomial Distribution**                            | **Normal Distribution**              |
| ----------------------- | ---------------------------------------------------- | ------------------------------------ |
| **Type**                | Discrete                                             | Continuous                           |
| **Values**              | Whole numbers (0, 1, 2, ...)                         | Any real number                      |
| **Shape**               | Symmetrical or Skewed                                | Bell-shaped (symmetrical)            |
| **Parameters**          | n = number of trials <br> p = probability of success | μ = mean <br> σ = standard deviation |
| **Example**             | Number of heads in coin tosses                       | People's height, weight              |
| **Approximation**       | Becomes like Normal when n is large                  | Always normal shape                  |
| **Formula for Mean**    | $\mu = np$                                           | Defined directly as μ                |
| **Formula for Std Dev** | $\sigma = \sqrt{np(1-p)}$                            | Defined directly as σ                |


```python
data = {
    "normal": random.normal(loc=50, scale=5, size=1000),
    "binomial": random.binomial(n=100, p=0.5, size=1000)
}
sns.displot(data, kind="kde")
plt.show()
```

## Poisson Distribution

```python
# Poisson Distribution is a Discrete Distribution.

# Estimates the probability of a number of events occurring within a fixed interval of time, space, or area, when events happen:
# Independently
# At a constant average rate

# Example:
# Number of calls received by a call center in 1 hour.
# If someone eats twice a day what is the probability he will eat thrice?

# It has two parameters:
# lam - rate or known number of occurrences e.g. 2 for above problem.
# size - The shape of the returned array.
```

```python
#  Generate a random 1x10 distribution for occurrence 2:

arr = random.poisson(lam = 2, size = 10)
print(arr)
```

## Visualization of Poisson Distribution

```python
sns.displot(random.poisson(lam = 2, size = 100))
plt.show()
```

# Difference Between Normal and Poisson Distribution

| Feature                       | **Normal Distribution**                                                                                         | **Poisson Distribution**                                                                                         |
| ----------------------------- | --------------------------------------------------------------------------------------------------------------- | ---------------------------------------------------------------------------------------------------------------- |
| **Type**                      | Continuous Distribution                                                                                         | Discrete Distribution                                                                                            |
| **Shape**                     | Symmetrical, Bell-shaped                                                                                        | Skewed (Right-skewed if mean is small), approximates normal when mean is large                                   |
| **Describes**                 | Continuous data like height, weight, IQ scores                                                                  | Count of events in a fixed time/space (e.g., calls per hour)                                                     |
| **Parameters**                | Mean (μ), Standard Deviation (σ)                                                                                | Rate (λ = average number of events)                                                                              |
| **Support (Possible Values)** | All real numbers (-∞ to ∞)                                                                                      | Non-negative integers (0,1,2,...)                                                                                |
| **Mean = Variance?**          | No (Variance = σ², independent of mean)                                                                         | Yes (Mean = Variance = λ)                                                                                        |
| **Example**                   | Heights of students, exam scores                                                                                | Number of emails received per hour, number of accidents per day                                                  |



```python
data = {
    "normal" : random.normal(loc = 50, scale = 7, size = 1000),
    "poisson" : random.poisson(lam = 50, size = 1000)
}
sns.displot(data, kind="kde")
plt.show()
```

#  Difference Between Binomial and Poisson Distribution

```python
# Binomial distribution only has two possible outcomes, whereas poisson distribution can have unlimited possible outcomes.
# But for very large n and near-zero p binomial distribution is near identical to poisson distribution such that n * p is nearly equal to lam.
```

```python
data = {
  "binomial": random.binomial(n=1000, p=0.01, size=1000),
  "poisson": random.poisson(lam=10, size=1000)
}
sns.displot(data, kind="kde")
plt.show()
```

## Logistic Distribution

```python
# A continuous probability distribution that looks like a bell-shaped curve (similar to Normal Distribution) but with sharper peaks and heavier tails.
# Logistic Distribution is used to describe growth.
# Used extensively in machine learning in logistic regression, neural networks etc.
```

| Parameter | Meaning                               | Default      |
| --------- | ------------------------------------- | ------------ |
| **loc**   | Mean (center/peak of the curve)       | 0            |
| **scale** | Controls **spread/flatness** of curve | 1            |
| **size**  | Number of data points to generate     | User-defined |


```python
#  Draw 2x3 samples from a logistic distribution with mean at 1 and stddev 2.0:

arr = random.logistic(loc=1, scale=2, size=(2,3))
print(arr)
```

## Visualization of Logistic Distribution

```python
sns.displot(random.logistic(size=1000), kind="kde")
plt.show()
```

## Difference Between Logistic and Normal Distribution

| **Point**          | **Normal Distribution**              | **Logistic Distribution**                                   |
| ------------------ | ------------------------------------ | ----------------------------------------------------------- |
| **Shape**          | Bell shape (smooth peak)             | Bell shape but **sharper peak**                             |
| **Tails**          | **Thin tails** (less extreme values) | **Thick tails** (more extreme values)                       |
| **Used In**        | General data like **height, weight** | **Machine learning** (Logistic Regression, Neural Networks) |
| **CDF**            | Complicated                          | Simple **Sigmoid function**                                 |
| **Spread Control** | Mean & Standard Deviation            | loc & scale                                                 |


```python
data = {
    "normal" : random.normal(scale = 2, size = 1000),
    "logistic" : random.logistic(size = 1000)
}
sns.displot(data, kind="kde")
plt.show()
```

##  Multinomial Distribution

```python
# An extension of the Binomial Distribution, but instead of just 2 outcomes (like heads/tails), you have more than 2 possible outcomes.

#  When to Use
# When you do an experiment n times, and each time there are k possible outcomes, 
# you use the Multinomial Distribution to find the probability of a specific combination of counts.

# Example
# Rolling a dice 5 times:
# Possible outcomes: 1, 2, 3, 4, 5, 6 (6 outcomes)
# Multinomial tells you the probability of getting:
# 1 twice
# 3 once
# 5 twice
# etc.

# It has three parameters:
# n - number of times to run the experiment.
# pvals - list of probabilties of outcomes (e.g. [1/6, 1/6, 1/6, 1/6, 1/6, 1/6] for dice roll).
# size - The shape of the returned array.
```

```python
#  Draw out a sample for dice roll:

arr = random.multinomial(n = 6, pvals = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6])
print(arr)
```

##  Exponential Distribution

```python
# Exponential distribution is used for describing time till next event e.g. failure/success etc.
# such as:
# Time until machine failure
# Time until a customer arrives
# Time between earthquakes
# Lifespan of electronic components
# Time until a phone call arrives in a call center

# It has two parameters:
# scale - inverse of rate ( see lam in poisson distribution ) defaults to 1.0.
# size - The shape of the returned array.
```

```python
#  Draw out a sample for exponential distribution with 2.0 scale with 2x3 size:

arr = random.exponential(scale = 2, size = (2,3))
print(arr)
```

```python
#  Visualization of Exponential Distribution

sns.displot(random.exponential (size = 1000), kind="kde")
plt.show()
```

# Relation Between Poisson and Exponential Distribution


### Poisson = "How many events in time T?"
### Exponential = "How long until next event?"

| **Aspect**       | **Poisson Distribution**            | **Exponential Distribution** |
| ---------------- | ----------------------------------- | ---------------------------- |
| **Type**         | Discrete                            | Continuous                   |
| **Deals With**   | Number of events in a time period   | Time between two events      |
| **Parameter**    | $\lambda$ (rate of events per time) | $\lambda$ (same as Poisson)  |
| **Example**      | Number of calls in 1 hour           | Time between two phone calls |
| **Key Property** | Counts multiple events              | Models the waiting time      |
| **Relation**     | Total events in time $t$            | Time gap between events      |


## Chi Square Distribution


```python
# Chi Square distribution is used as a basis to verify the hypothesis.
# A hypothesis is just a guess or assumption about something.

# The Chi-Square Distribution is used to:
# Check whether your guess (hypothesis) is correct or not.
# It compares actual data vs expected data.

# It has two parameters:
# df - (degree of freedom).
# size - The shape of the returned array.
```

```python
#  Draw out a sample for chi squared distribution with degree of freedom 2 with size 2x3:

arr = random.chisquare(df=2, size=(2, 3))
print(arr)
```

## Visualization of Chi Square Distribution

```python
sns.displot(random.chisquare(df = 1, size = 1000), kind = "kde")
plt.show()
```

## Rayleigh Distribution

```python
# Rayleigh distribution is used in signal processing.
# Where It's Used
# Signal Processing: To model noise or fading signals.
# Wind Speed Analysis: To model wind speeds.
# Engineering & Physics: In scenarios involving random vibrations or scattering.

# Example :-
# You are studying wind speeds in an open field.
# The wind blows in random horizontal (x) and vertical (y) directions.
# Both directions follow a normal distribution.
# You want to know:
# What is the probability of a particular wind speed (magnitude)?
# Here, the Rayleigh distribution models the overall wind speed (magnitude).

# It has two parameters:
# scale - (standard deviation) decides how flat the distribution will be default 1.0).
# size - The shape of the returned array.
```

```python
#  Draw out a sample for rayleigh distribution with scale of 2 with size 2x3:

arr = random.rayleigh(scale = 2, size = (2,3))
print(arr)
```

## Visualization of Rayleigh Distribution

```python
sns.displot(random.rayleigh(size = 1000), kind = "kde")
plt.show()
```

## Similarity Between Rayleigh and Chi Square Distribution
#### At unit stddev and 2 degrees of freedom rayleigh and chi square represent the same distributions.



## Pareto Distribution


```python
# A distribution following Pareto's law i.e. 80-20 distribution (20% factors cause 80% outcome)
# Real-Life Examples of Pareto Distribution :-
# Social Media -	20% of users create 80% of the content.
# Traffic Accidents -	20% of roads see 80% of accidents.

# It has two parameter:
# a - shape parameter.
# size - The shape of the returned array.
```

```python
#  Draw out a sample for pareto distribution with shape of 2 with size 2x3:

arr = random.pareto(a = 2, size = (2,3))
print(arr)
```

##  Visualization of Pareto Distribution

```python
sns.displot(random.pareto(a=2, size = 1000))
plt.show()
```

##  Zipf Distribution

```python
# Zipf distributions are used to sample data based on zipf's law.
#  Zipf's Law :- As the rank increases, the frequency decreases sharply.

# | **Rank** | **Word** | **Frequency (Approx.)** |
# | -------- | -------- | ----------------------- |
# | 1        | "the"    | 5000 times              |
# | 2        | "of"     | 2500 times              |
# | 3        | "and"    | 1666 times              |
# | 4        | "to"     | 1250 times              |
# | ...      | ...      | ...                     |



# It has two parameters:
# a - distribution parameter.
# size - The shape of the returned array.
```

```python
#  Draw out a sample for zipf distribution with distribution parameter 2 with size 2x3:

arr = random.zipf(a = 2, size = (2,3))
print(arr)
```

##  Visualization of Zipf Distribution

```python
#   Sample 1000 points but plotting only ones with value < 10 for more meaningful chart.

x = random.zipf(a=2, size=1000)
sns.displot(x[x<10])
plt.show()
```
